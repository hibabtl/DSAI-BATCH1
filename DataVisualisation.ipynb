```python
#Data Analysis and Visualization
KMs_Driven_counts=data['KMs Driven'].value_counts().head(6)
plt.pie(KMs_Driven_counts,labels=KMs_Driven_counts.index, autopct='%1.1f%%', radius=2, startangle=90, pctdistance=0.8)
plt.axis('equal')
plt.title('DISTRIBUTION OF KMs Driven')

plt.show()
brand_counts = data.groupby(['Registered City', 'Brand']).size().unstack()

# Plot the bar chart
brand_counts.plot(kind='bar', stacked=True, figsize=(10, 10))

# Set labels and title
plt.xlabel('Registered City')
plt.ylabel('Number of Cars')
plt.title('Car Brands in Registered Cities')

# Show the legend
plt.legend(title='Brand')

# Display the chart
plt.show()

```
